## Prediction A and B

```python
# Predict and save results
def predict_surprise_storms_task1(file_path, event_ids, output_dir, team_name):
    task_1.os.makedirs(output_dir, exist_ok=True)

    for event_id in event_ids:
        print(f"Processing {event_id}...")

        # Load data
        vis, ir069, ir107, vil = task_1.task1_load_multichannel_data(file_path, event_id)

        # Task 1A: only 'vil'
        X_1A = vil[..., task_1.np.newaxis]  # (192, 192, 12, 1)
        predicted_1A = saved_a.predict(task_1.np.expand_dims(X_1A, axis=0))[0]
        # predict

        # Task 1B: use 'vis, ir069, ir107, vil'
        X_1B = task_1.np.stack([ir069, ir107, vil, vis], axis=-1)
        # (192, 192, 12, 4)
        predicted_1B = saved_b.predict(task_1.np.expand_dims(X_1B, axis=0))[0]

        predicted_1A_resized = task_1.np.array([task_1.cv2.resize(predicted_1A[:, :, i], (384, 384), interpolation=task_1.cv2.INTER_CUBIC)
                                 for i in range(12)]).transpose(1, 2, 0).astype(task_1.np.float32)

        predicted_1B_resized = task_1.np.array([task_1.cv2.resize(predicted_1B[:, :, i], (384, 384), interpolation=task_1.cv2.INTER_CUBIC)
                                 for i in range(12)]).transpose(1, 2, 0).astype(task_1.np.float32)
        # Save the results
        task_1.np.save(task_1.os.path.join(output_dir,
                                           f"{team_name}-task1a-vil-{event_id}.npy"), predicted_1A_resized)
        task_1.np.save(task_1.os.path.join(output_dir,
                                           f"{team_name}-task1b-vil-{event_id}.npy"), predicted_1B_resized)

        print(f"Saved predictions for {event_id}")
```

```python
# Parameters
h5_file = "data/surprise_task1.h5"
csv_file = "data/surprise_events1.csv"
output_dir = "predictions"
team_name = "barry"

# Load event IDs
event_ids = load_event_ids(csv_file)

# Run prediction
predict_surprise_storms_task1(h5_file, event_ids, output_dir, team_name)
```

### Task1 A and B the result(gif)

```python
def make_gif(files, fps=10):
    """
    Helper function to create a GIF from a sequence of images.
    """
    imgs = [task_1.PIL.Image.open(file) for file in files]
    gif_path = "_temp.gif"
    imgs[0].save(fp=gif_path, format='GIF', append_images=imgs[1:],
                 save_all=True, duration=int(1000/fps), loop=0)
    return task_1.IPython.display.Image(filename=gif_path)


def plot_event(event_id, delete_temp=True):
    """
    Generate side-by-side GIFs for input and predictions.
    """
    # Load data
    vis, ir069, ir107, vil = task_1.task1_load_multichannel_data(h5_file, event_id)

    # Read the predictions

    pred_task1a = task_1.np.load(task_1.os.path.join(output_dir, f"barry-task1a-vil-{event_id}.npy"))
    # Task 1A Prediction
    pred_task1b = task_1.np.load(task_1.os.path.join(output_dir, f"barry-task1b-vil-{event_id}.npy"))
    # Task 1B Prediction

    # Last 12 frames as input
    vil_input = vil[:, :, :12]

    # Create and save individual frames
    temp_files_1a = []
    temp_files_1b = []

    for i in range(vil_input.shape[2]):  # 12 frames
        fig, axs = task_1.plt.subplots(1, 2, figsize=(10, 4))
        axs[0].imshow(vil_input[:, :, i], cmap="turbo")  # input
        axs[1].imshow(pred_task1a[:, :, i], cmap="turbo")  # predict
        axs[0].set_title(f"{event_id} - Input Frame {i+1}")
        axs[1].set_title(f"{event_id} - Task 1A Prediction Frame {i+1}")
        axs[0].axis("off")
        axs[1].axis("off")

        temp_file = f"_temp_{event_id}_1a_{i}.png"
        fig.savefig(temp_file, bbox_inches="tight", dpi=100, pad_inches=0.02, facecolor="white")
        task_1.plt.close(fig)
        temp_files_1a.append(temp_file)

        # Task 1B
        fig, axs = task_1.plt.subplots(1, 2, figsize=(10, 4))
        axs[0].imshow(vil_input[:, :, i], cmap="turbo")
        axs[1].imshow(pred_task1b[:, :, i], cmap="turbo")
        axs[0].set_title(f"{event_id} - Input Frame {i+1}")
        axs[1].set_title(f"{event_id} - Task 1B Prediction Frame {i+1}")
        axs[0].axis("off")
        axs[1].axis("off")

        temp_file = f"_temp_{event_id}_1b_{i}.png"
        fig.savefig(temp_file, bbox_inches="tight", dpi=100, pad_inches=0.02, facecolor="white")
        task_1.plt.close(fig)
        temp_files_1b.append(temp_file)

    # Create GIFs and display
    print(f"Displaying GIFs for {event_id}")
    display(make_gif(temp_files_1a, fps=5))  # Task 1A
    display(make_gif(temp_files_1b, fps=5))  # Task 1B

    # Cleanup temporary files
    if delete_temp:
        for file in temp_files_1a + temp_files_1b:
            task_1.os.remove(file)
```

```python
event_ids = load_event_ids(csv_file)

for event_id in event_ids:
    plot_event(event_id)
```
